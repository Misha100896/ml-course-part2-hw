```python
import datetime as dt
 
def myARMA(x, h, Params):
    p = Params['p']
    q = Params['q']
   
    start = x.index[-1] + dt.timedelta(days=1)
    end = x.index[-1] + dt.timedelta(days=h)
    rng = pd.date_range(start, end)
    frc_ts = pd.Series(index=x.index.append(rng))
 
    for t in range(len(x.index)):
        y = x.iloc[:t + 1]
       
        if not y.dropna().empty:
            try:
                arma_mod = sm.tsa.ARMA(y.dropna(), (p, q)).fit()
                frc_ts[t + h] = arma_mod.forecast(steps=h)[0][0]
            except Exception as exc:
                pass
 
    return frc_ts.ffill()
```

```python
ARMA = myARMA(data_test, h=1, Params={'p':1, 'q':1})
```

```python
%%time
best_aic = float('inf')
best_params = ()
for p in range(1, 6):
    for q in range(1, 4):
        for d in range(1, 2):
            arima = ARIMA(data_train, order=(p, d, q)).fit()
            if arima.aic < best_aic:
                best_aic = arima.aic
                best_params = {'p': p, 'd': d, 'q': q}
```

```python
print (best_aic)
print (best_params)
```

```python
import datetime as dt
 
def myARIMA(x, h, Params):
    p = Params['p']
    q = Params['q']
   
    start = x.index[-1] + dt.timedelta(days=1)
    end = x.index[-1] + dt.timedelta(days=h)
    rng = pd.date_range(start, end)
    frc_ts = pd.Series(index=x.index.append(rng))
 
    for t in tqdm(range(len(x.index))):
        y = x.iloc[:t + 1]
       
        if not y.dropna().empty:
            try:
                arma_mod = sm.tsa.ARMA(y.dropna(), (p, q)).fit()
                frc_ts[t + h] = arma_mod.forecast(steps=h)[0][0]
            except Exception as exc:
                pass
 
    return frc_ts.ffill()
```

```python
prediction = myARIMA(data[50:], 1, Params=best_params)
```

```python
prediction.plot(figsize=(12,7))
data.plot(figsize=(12,7))
plt.plot()
```

```python
print('MACAPE: %.2f'% qualityMACAPE(data[-285:], prediction)[0])
print('RMSE: %.2f'% qualityRMSE(data[-285:], prediction)[0])
```
